```python
import os
from dotenv import load_dotenv
import requests
from openai import OpenAI
from IPython.display import Markdown
```

```python
load_dotenv(override = True)
api_key = os.getenv("OPENAI_API_KEY")

if not api_key:
    print("No API key was found - please head over to the troubleshooting notebook in this folder to identify & fix!")
elif not api_key.startswith ("sk-proj-"):
    print("An API key was found, but it doesn'tstart with sk-proj-; please check you're using the right key - seetroubleshooting notebook")
else:
    print("API key found and looks good so far!")
```

```python
headers = {"Authorization": f"Bearer {api_key}", "Content-Type": "application/json"}

payload = {
    "model": "gpt-5-nano",  # Note: "nano" with the letter o — not "nan0" with a zero
    "messages": [
        {"role": "user", "content": "Tell me a fun fact"}
    ]
}

payload
```

```python
response = requests.post(
    "https://api.openai.com/v1/chat/completions",
    headers = headers,
    json = payload
)

response.json()
```

```python
response.json()["choices"][0]["message"]["content"]
```

```python
openai = OpenAI()

response = openai.chat.completions.create(model = "gpt-5-nano", 
                                          messages = [{"role": "user",
                                          "content": "Tell me a fun fact"}])
response.choices[0].message.content                                        
    
```

```python
GEMINI_BASE_URL = "https://generativelanguage.googleapis.com/v1beta/openai/"  # Must be "openai" — not "openal"

load_dotenv(override=True)

google_api_key = os.getenv("GOOGLE_API_KEY")

if not google_api_key:
    print("No API key was found - please be sure to add your key to the .env file, and save the file! Or you can skip the next 2 cells if you don't want to use Gemini")
elif not google_api_key.startswith("AIza"):
    print("An API key was found, but it doesn't look like a Google AI Studio key.")
    print(f"  Expected prefix: AIza  |  Your key starts with: {google_api_key[:4]}")
    print("  Get a key at: https://aistudio.google.com/api-keys")
    print("  Then add to .env: GOOGLE_API_KEY=AIza...")
else:
    print("API key found and looks good so far!")
    
```

```python
gemini = OpenAI(base_url=GEMINI_BASE_URL, api_key=google_api_key)

response = gemini.chat.completions.create(
    model="gemini-3.5-flash",  # Plain string — no backticks! (matches day2.ipynb)
    messages=[{"role": "user", "content": "Tell me a fun fact"}]
)
response.choices[0].message.content
```
